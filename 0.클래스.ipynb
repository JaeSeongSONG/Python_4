# class

- 개념
  - entity (개체)
    - 이 세상에 존재하는 모든 개념(무형), 유형(실존하는것)  
    - 이런 entity를 컴퓨터 사이언스, 프로그래밍 세상에 entity를 표현하기 위해서는 도구가 필요 => class라는 문법(표현도구)를 가지고 구현
  - class
    - 문법이 존재, 엔티티를 표현하는 도구
    - 클레스를 작성 -> 실행 -> 클레스 생성자를 호출
    - 클레스의 모습을 본딴 형태로 메모리에 뭔가가 상주하게 된다 => 객체(object)
    - 클레스의 인스턴스를 객체라고 부른다
      - 여러개 만들수 있다
    - 객체의 청사진은 클레스라고 부른다

  - object
    - 프로그램이 작동하면 메모리상에 존재
    - 메모리 상에 존재하니까, 주소가 존재 (0x124123sdfds)
    - 특정 변수가 그 객체를 받아주면 이제부터 그 변수를 xxx 객체라고 부른다
    - 정확하게는 특정 변수가 주소 0x124123sdfds를 가진 객체를 가르킨다 (특정 객체를 가리키는 주소를 가진다)

# 클레스 구성원

- 생성자
- 맴버
  - 변수
    - 접근제어를 표현할수 있는데 언급(차후)
  - 함수
- 소멸자 (차후)
- 기타 _로 시작하는 사전 정의된 함수들(차후)
- 정적함수 (차후)


```python
# 맴버 -> 클레스 소속이다!! -> 객체명.맴버변수 or 객체명.맴버함수
# Person 클레스 내부에서 자기 자신을 표현할때 => self or(타 언어는 this)

# 부모 클레스를 명시적으로 표현하지 않아도 이미 Object(모든 클레스의 수퍼클레스)를 상속하고 있다
#class Person(Object):
class Person:
  # 생성자 (생략 가능함 -> 부모 생성자가 있으니까? 그것을 그대로 사용)
  # 맴버 변수 
  p = 0.2
  # 맴버 함수  
  def getP(self, a):
    return self.p
  
per = Person()
```

```python
#객체명.맴버변수
per.p
```

```python
#객체명.맴버함수
per.getP()
```

# 상속

- 부모 클레스로부터 모든 유산를 받아서 자식 클레스는 사용 가능
- 부모로 부터 받은 유산을 재정의해서 사용 가능
- 자식 클레스는 별로도 새로운 기능을 만들어서 추가할수 있다
- XMan > Person > Object
  - XMan은 Object 이다 ( is a )
  - Object는 자식으로 XMan을 가진다 ( has a ) 

```python
class XMan(Person):
  def attack(self):
    print( '하늘을 나는 능력을 추가했다' )

x = XMan()
```

```python
x.p, x.getP(), x.attack()
```
